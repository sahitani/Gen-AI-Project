```python
# Cell 3 — Your first Claude API call
import anthropic

client = anthropic.Anthropic(api_key=ANTHROPIC_API_KEY)

response = client.messages.create(
    model="claude-sonnet-4-6",
    max_tokens=512,
    system="You are an AI Programme Assistant who helps with enterprise AI delivery.",
    messages=[
        {"role": "user", "content": "What are the top 3 risks in a GenAI enterprise deployment?"}
    ]
)

reply = response.content[0].text
print("Claude says:\n")
print(reply)
```

```python
# Cell 5 — Compare prompts using Claude only
def ask_claude(question, system_prompt="You are a helpful AI assistant."):
    response = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=512,
        system=system_prompt,
        messages=[{"role": "user", "content": question}]
    )
    return response.content[0].text

# Ask Claude the same question with two different system prompts
question = "What should a Programme Manager focus on before deploying a GenAI solution?"

print("=" * 50)
print("CLAUDE as a TECHNICAL EXPERT:")
print("=" * 50)
print(ask_claude(question, system_prompt="You are a technical GenAI architect with 10 years of experience."))

print("\n" + "=" * 50)
print("CLAUDE as a BUSINESS CONSULTANT:")
print("=" * 50)
print(ask_claude(question, system_prompt="You are a business consultant focused on ROI and stakeholder management."))
```

```python
# Cell 6 — Stateful Chatbot that remembers conversation history
class ChatBot:
    def __init__(self, system_prompt="You are a helpful AI assistant."):
        self.system_prompt = system_prompt
        self.history = []  # this empty list is the memory

    def chat(self, user_message):
        # Step 1 — add user message to history
        self.history.append({"role": "user", "content": user_message})

        # Step 2 — send full history to Claude every time
        response = client.messages.create(
            model="claude-sonnet-4-6",
            max_tokens=512,
            system=self.system_prompt,
            messages=self.history
        )

        reply = response.content[0].text

        # Step 3 — add Claude's reply to history too
        self.history.append({"role": "assistant", "content": reply})

        return reply

    def show_history(self):
        for msg in self.history:
            print(f"[{msg['role'].upper()}]: {msg['content']}\n")


# --- Test it ---
bot = ChatBot(
    system_prompt="You are an AI Programme Assistant specialising in enterprise GenAI delivery."
)

# Turn 1
print("Turn 1:")
print(bot.chat("I am managing a GenAI programme for a large retail client. What should I focus on first?"))

# Turn 2
print("\nTurn 2:")
print(bot.chat("What about data privacy risks specifically for retail?"))

# Turn 3
print("\nTurn 3:")
print(bot.chat("Give me a one line summary of everything we just discussed."))

# See full memory
print("\n--- Full conversation history stored in memory ---")
bot.show_history()
```
